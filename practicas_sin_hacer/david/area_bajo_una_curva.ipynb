AREA BAJO UNA CURVA

```python
from sympy.abc import x
import numpy as np
from matplotlib import pyplot as plt
from sympy import lambdify
puntoinicial =0 # Punto inicial de la curva
puntofinal=6 # Punto final de la curva
numerocortes=1000 # Número de cortes de la curva

funcion=cos(2*pi*x)*exp(-x)+1 # Función a evaluar

xvalues=np.linspace(puntoinicial,puntofinal,numerocortes) # Valores de x
yvalues=lambdify(x,funcion)(xvalues) # Valores de y

plt.plot(xvalues,yvalues) # Gráfico de la función

#facecolor='color'
#coordenadaY= coordenada Y apartir de de la que se rellena el gráfico 
#alpha=opcidad
#plt.fill_between(xvalues,yvalues,coordenadaY,where,facecolor,alpha)

plt.fill_between(xvalues,yvalues,0.25,where=(xvalues>=1)&(xvalues<=4),facecolor='blue',alpha=0.2) # Relleno de la función

#plt.text(x,y,texto,fontsize=tamaño,color=color,rotation=angulo)

plt.text(2.5,0.4,r"$\int_{1}^4(cos(2πx)e^{-x}+1)\mathrm{d}x$",horizontalalignment='center',fontsize=14)
plt.show()
```
