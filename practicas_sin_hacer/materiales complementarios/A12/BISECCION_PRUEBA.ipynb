```python
# ALGORITMO DE BISECCIÃ“N

import numpy as np
from sympy import *

# ENTRADA

f = lambda x: x**2-1
a = 0
b = 3
tolerancia = 0.001

f = lambda x: x**2-sin(x)-0.5
a = float(0)
b = float(2)
tolerancia = 0.0001


# PROCEDIMIENTO

tabla = []
h = b - a

fa = f(a)
fb = f(b)
i = 1

while (h > tolerancia):
    h = h / 2
    c = (a + b) / 2
    fc = f(c)
    tabla.append([i,a,c,b,fa,fc,fb,h])
    i = i + 1
                 
    signo = np.sign(fa)*np.sign(fc)
    if (signo < 0):
        b = c
        fb = fc
    else:
        a = c
        fa = fc
 
c = (a + b) / 2
fc = f(c)
tabla.append([i,a,c,b,fa,fc,fb,h])
tabla = np.array(tabla)

raiz = c

# SALIDA

print(' i          a            c            b          f(a)         f(c)        f(b)           h')
n = len(tabla)

for i in range(0,n,1):
    unafila = tabla[i]
    #formato = '{:>2.0f}'+'   '+(len(unafila)-1)*'{:>10.6f}   '
    formato = '{:0>2d}'+'   '+(len(unafila)-1)*'{:>10.6f}   '
    unafila = formato.format(*unafila)
    print(unafila)  
print('\n','raiz: ',raiz,'\n','\n')

for i in range(0,n,1):
    unafila = tabla[i]
    formato = '{!s:>2.2s}'+' '+(len(unafila)-1)*'{!s:>12.9s} '
    unafila = formato.format(*unafila)
    print(unafila) 
print('\n','raiz: ',raiz,'\n','\n')
```
